<h1><center>EDA on Dataset</center></h1>

```python
import pandas as pd
import numpy as np 

df=pd.read_csv('data/Data_Train.csv') 
df.head()
```

```python
df.info()  ## information of dataset
```

<p>Dataset have 11 features in which 10 are object type and 1 is integer type (which is target column also). </p>

```python
df.shape  ## dataset have 13354 rows and 11 columns.
```

```python
df.isnull().sum()    ## since there is not much null values so we can simply drop it.
```

```python
df.dropna(inplace=True)   ## drop the null/na values.
df.isnull().sum()  ## checking again for value count.
```

<h3>Converting objects type features into integer values.</h3>

```python
df['Date_of_Journey']
```

```python
df['day_of_journey']=pd.to_datetime(df['Date_of_Journey'], format="%d/%m/%Y" ).dt.day   ## making day_of_journey column
df['month_of_journey']=pd.to_datetime(df['Date_of_Journey'], format="%d/%m/%Y" ).dt.month  ## making month_of_journey column


df.head()
```

<h4>Now we coverted date of journey in day and month columns, we can drop the date of journey column.</h4>

```python
df.drop(['Date_of_Journey'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Departure time is when a plane leaves the gate. 
# Similar to Date_of_Journey we can extract values from Dep_Time

# Extracting Hours
df["Dep_hour"] = pd.to_datetime(df["Dep_Time"]).dt.hour

# Extracting Minutes
df["Dep_min"] = pd.to_datetime(df["Dep_Time"]).dt.minute

# Now we can drop Dep_Time as it is of no use
df.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
# Arrival time is when the plane pulls up to the gate.
# Similar to Date_of_Journey we can extract values from Arrival_Time

# Extracting Hours
df["Arrival_hour"] = pd.to_datetime(df.Arrival_Time).dt.hour

# Extracting Minutes
df["Arrival_min"] = pd.to_datetime(df.Arrival_Time).dt.minute

# Now we can drop Arrival_Time as it is of no use
df.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
## we have a feature name duration since there is no direct connection with the target feature. Duration is the difference of date of jounrey and arrival time.
## it cannot take as a input from user, so we can drop it

df.drop('Duration', axis=1, inplace=True)
```

```python
## similarly Route and total_stop both are related to each other,
## And 80% of Additional_Info have no info,
## So we drop both the column route and Additional_Info.

df.drop(['Additional_Info', 'Route'], axis=1, inplace=True)
```

```python
df.head()
```

<h4>We clean all the data now we handle the catagorical date (columns).</h4>
<h4>Airline, Source, Destination, Total_Stops are the catagorical columns.</h4>

```python
## checking the value count of each catagorical columns.

cat_col=['Airline', 'Source', 'Destination', 'Total_Stops']

for cat in cat_col:
   
    print(df[cat].value_counts())
    print('===============================================')
    print('\n')
```

<h4>From the above observation</h4>
<p>Airline , Source, Destination are to be encoded bt <b>OneHotEncoder</b> and Total_Stop to be <strong>LabelEncoder</strong>.</p>

```python
## Applying OneHotEncodeing on Aireline

Airline = df[["Airline"]]

airline = pd.get_dummies(Airline, drop_first= True)

airline.head()
```

```python
## Applying OneHotEncodeing on Source

Source = df[["Source"]]

source = pd.get_dummies(Source, drop_first= True)

source.head()
```

```python
## Applying OneHotEncodeing on Destination

Destination = df[["Destination"]]

destination = pd.get_dummies(Destination, drop_first= True)

destination.head()
```

```python
# As this is case of Ordinal Categorical type we perform LabelEncoder
# Here Values are assigned with corresponding keys


df.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
df.head()
```

```python
# Concatenate dataframe --> train_data + Airline + Source + Destination

df = pd.concat([df, airline, source, destination], axis = 1)
df.head()
```

```python
## Now we have to drop Airline, Source, Destination

df.drop(['Airline', 'Source', 'Destination'], axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
## saving the finall dataframe in csv file

df.to_csv("data/train_data_cleaned.csv", index=None)
```
